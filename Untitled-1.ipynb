4.Limpieza de Datos previo al análisis

```python
df_sinduplicados = df.drop_duplicates()
print("Tamaño del DF Original:", len(df))
print("Tamaño del DF sin duplicados:", len(df_sinduplicados))
print("Registros/Tuplas eliminadas:", len(df) - len(df_sinduplicados))
```

Importante: La eliminación de duplicaods debe realizarse con extremo cuidadosobre todo con datos agrupadoso que no tienen una clase primaria o atributo identificador, porque existe el reisgo
de alterar los resltados y tomar decisiones.

```python
# Eliminar registros con algún dato vacío en cualquier columna
df_sinvacios = df.dropna()

print("Tamaño del DF Original:", len(df))
print("Tamaño del DF sin datos nulos:", len(df_sinvacios))
print("Registros/Tuplas eliminadas:", len(df) - len(df_sinvacios))
```

5.Definir los criterios de análisis (categorias)

```python
analysis_criteria = [
    'work_year',
    'experience_level',
    'employment_type',
    'job_title',
    'salary',
    'salary_currency',
    'salary_in_usd',
    'employee_residence',
    'remote_ratio',
    'company_location',
    'company_size'
]
```

6.Graficamos algo Basico

```python
def_color = 'darkblue'

for f in analysis_criteria:
    plt.figure(figsize=(12, 3))
    df[f].value_counts().plot(kind='bar', color=def_color)
    plt.title(f)
    plt.grid()
    plt.show()
```

6.Limpiamos los dtos y solo nos enfocamos en los datos recolectados desde 2023

```python
# Creamos un subconjunto de la muestra original
df_after2023 = df[
    (df['work_year'] >= 2023) &
    (df['employment_type'] == 'FT') &
    (df['company_location'] == 'US')
]

df_after2023.info()
df_after2023.size

```

```python
counts = df["work_year"].value_counts().sort_index()
ax = counts.plot(kind="bar", figsize=(10, 5), color='skyblue')
# Agregar los totales encima
for i, v in enumerate(counts):
    ax.text(i, v + 500, str(v), ha='center', fontweight='bold')
plt.title("Total por año")
plt.xlabel("Año")
plt.ylabel("Cantidad")
plt.grid(axis='y')
plt.show()
```

```python
#Mostrar la distribucion del trabajo remoto
plt.figure(figsize=(7,3))
df_after2023.remote_ratio.value_counts().sort_index().plot(kind='bar', color=def_color)
plt.title('Distribucion de modalidad de trabajo para las empresas estadounidenses')
plt.grid()
plt.show()
```
